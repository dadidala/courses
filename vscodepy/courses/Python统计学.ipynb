**一些统计知识**
* 关于带宽对核密度估计曲线（--kernel density estimation）的影响  
带宽过大，核密度估计曲线会变得更加平滑。这是因为较大的带宽意味着在估计某一点的密度时，考虑了更广泛范围内的数据点，从而对局部的波动起到了更强的平滑作用；可能会掩盖数据的一些局部特征。由于平滑作用过强，一些数据中的细微结构可能会被忽略。比如，数据中可能存在的多个峰值在较大带宽下可能会被合并成一个较宽的峰，从而失去了对数据多模态特征的准确刻画  
核密度估计曲线会更加接近数据的实际分布情况，呈现出更多的细节。较小的带宽使得在估计密度时只考虑了邻近的数据点，能够更好地捕捉到数据的局部变化；但同时也会使估计结果变得更加粗糙，出现更多的波动。因为只依赖少数邻近的数据点进行估计，容易受到个别数据点的影响，导致估计结果不稳定。

```python
#使用GridSpec函数布局
import matplotlib. pyplot as plt
fig = plt. figure (figsize=(6,5))  #设置图形大小
grid=plt. GridSpec (3, 3)          #生成3*3网格
plt. subplot (grid[0, :2])         #占据第一行,前两列
plt. subplot(grid[0, 2])           #占据第一行,第三列
plt.subplot(grid[1,:1])            #占据第二行,第一列
plt.subplot(grid[1, 1:])           #占据第二行,后两列
plt.subplot(grid[2,:3])            #占据第三行前三列
fig. tight_layout                  #紧凑布局
```

```python
#使用subplot2grid函数布局
import matplotlib. pyplot as plt
fig = plt. figure(figsize=(6, 4.5))
ax1=plt.subplot2grid((3,3),(0,0),rowspan=2,colspan=2)    #第一行第一幅图,占据两行两列
ax2 =plt.subplot2grid((3,3), (0,2),rowspan=3,colspan=1)  #第一行第二幅图,占据三行一列
ax3 =plt.subplot2grid((3,3), (2,0),rowspan=1,colspan=1)  #第三行第一幅图,占据一行一列
ax4 =plt. subplot2grid((3,3),(2,1),rowspan=1,colspan=1)  #第三行第二幅图,占据一行一列
fig. tight_layout ()
```

```python
#使用add_gridspec函数布局
import matplotlib. pyplot as plt
import numpy as np
np. random. seed (1010)
fig =plt. figure(figsize=(7, 4))
spec =fig. add_gridspec (nrows=2, ncols=6, width_ratios=[1,1,1,1,1,1],height_ratios=[1,2])






fig. add_subplot (spec [0, 1:3])         #占据第1行的2日3列
fig. add_subplot (spec [0, 3:])          #占据第1行的后三列
ax =fig. add_subplot (spec [:, 0])       #占据第1列的所有行
ax =fig. add_subplot (spec [1, 1:4])     #占据第2行的2-4列
ax =fig. add_subplot (spec [1, 4:])      #占据第2行第4列后面的所有列
fig. tight_layout ()
plt. show ()
```
