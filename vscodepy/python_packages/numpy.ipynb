* 关于对切片的解读a[a:b],解读为从a开始到(b-1)结束

# 关于筛选特定数据

补充一下上述网址没包含位运算符，即& | ~   
两个布尔数组进行 & 运算时，会进行逐位比较，并返回一个新的布尔数组。对于每一对位置上的布尔值，& 运算符将按以下规则计算  
* True&True==>True
* False&True==>False
* True&False==>False
* False&False==>False  
"a<n"会返回一个**索引掩码**，[True,True,False,False,True],然后索引掩码再带入a中，即a[np.array([True,True,False,False,True])],则只返回为True的对应位置的元素，即返回含三个元素的数组[a[0],a[1],a[4]];直接用列表也可以，即a[[True,True,False,False,True]]  
numpy通过位运算符组合不同的条件，&是按位与运算符，用来在两个布尔数组之间进行逐位比较，最终得到一个新的布尔数组。  
a=np.arange(10)  
(a < 8) & (a > 3) 会对 a < 8 和 a > 3 的结果逐位进行 "**与**" 运算，返回一个新的布尔数组。  
(a < 8)  # [True, True, True, True, True, True, True, True, False, False] ---(1)  
(a > 3)  # [False, False, False, False, True, True, True, True, True, True ---(2)  
(1)式和(2)式通过逐位进行"与"逻辑运算，得出一个新的波尔数组，逐位是指(1)式的第$i$和(2)式的第$i$位进行**与**逻辑操作  
[False, False, False, False, True, True, True, True, False, False]  
mask=(a<8>)&(a>3)  
a[mask] #返回的是一个满足条件值，形状为一维数组  
np.where(mask,a,0) #返回的是跟原数组相同形状的矩阵，where函数的第一个参数是筛选条件（即该参数输入索引掩码），第二个参数是从哪里筛选，第三个参数是在False（即不满足条件）的位置显示指定内容，也可以是np.nan  
np.argwhere(mask) #返回满足条件的索引，若$a_{ij}$满足条件，则会以数组形式返回其下标[i,j]

# 有关于维度的解释

有关于矩阵的维度的相关解释：  
* 关于维度的看法是：从最后一位向前看 ，列，行，层... 
* axis = 0，axis = 1...；其中0,1,2,3...n；0是指形状的第一个元素，相当于将形状看成一个序列（从零开始计数），axis=n就是序列的索引，是指（$D_1,D_2,D_3,...,D_i$）中的第n个元素，例如：对于一个四维矩阵(2,4,3,5)，2对应着最高维四维（由两个三维矩阵构成）axis=0，是指第一个元素也就是第四维，axis=2是指第二维，同理axis=3是指第一维
* （1,2,2,1,4,5） 这个维度看起来是六维矩阵，但是其中的“1”维度可以省去，即该维度视为四维矩阵
* 其中一维维度有五个元素，即一个一维有五个元素；二维维度有四个元素，即一个二维矩阵由四个一维矩阵构成；三维维度有两个元素，即一个三维矩阵由两个二维矩阵构成；一个四维维度有两个元素，即一个四维矩阵由两个三维矩阵构成；
* 一维由n个元素构成，二维由n个一维构成，三维由n个二维构成，n维由n-1维构成,即3D=2D+2D+2D+...+2D;4D=3D+3D+3D+...+3D;
* 最高维的个数永远是一，即四维矩阵的最高维是四维，且只有一个四维矩阵，其他维度矩阵的个数可以是任意的

* 对矩阵的二维进行聚类函数等函数操作，其作用的的矩阵的当前维度-1的矩阵，如对一维进行.max（）,[1,2,3,4,5,6,7,8,9].max() ==>[9],其作用的矩阵就是单个元素，即每个元素进行比较求出max；同理对二维矩阵进行函数操作，其作用的矩阵是2-1维的矩阵，即每个一位进行max比较，最后返回的结果就是一个一维数组；对三维进行函数操作，即各个二维矩阵间进行比较，最后返回的结果是一个二维数组，即 MAX 3D =MAX[2D 2D 2D ... 2D] ==>2D,即MAX 3D ==> 2D，对三维求max会返回一个二维的结果;MAX 4D=MAX[3D 3D 3D ...3D] ==> 3D，对四维求MAX返回一个三维结果；

* 还有就是对四维矩阵np.random.rand(2,4,3,5)的三维进行.max(axis=1)，按照上一条的规律总结，对三维进行MAX之后其作用的3-1维矩阵，也就是二维矩阵，MAX 3D==> 2D,即返回一个二维矩阵；又因为4D = [3D 3D 3D ...3D];[3D 3D 3D ...3D]==>[2D 2D 2D ... 2D];故4D = [2D 2D 2D ... 2D] ，即4D由多个2D构成，但是也可反正看，由多个2D构成的矩阵是3D，故此时4D降维成3D,3D降维成2D，即对n维进行函数操作之后，n维和比n维更高的维的维度都-1；还可以换一种角度看，就是对于四维(2,4,3,5)中的三维进行降维打击即四维(2,4,3,5)==>(2,1,3,5)==>(2,3,5),即一维二维的元素个数不变，三维的个数变为一
* 比如关于对三维求.max（axis=1）的作用机理，如果对于一个四维矩阵np.random.rand(2,4,3,5)对三维进行.max(axis=1)，其作用的是二维矩阵，即每个三维矩阵中的n个二维进行比较然后返回最大值的二维矩阵，这样每个三维矩阵是由一个二维矩阵构成，那么一个四维（最高维）所包裹的m个三维就变成了四维包裹的m个二维，由m个二维所构成的是三维，即此时四维转变为三维，原因是原有的三维的元素个数被降为一个相当于(2,4,3,5)~~>(2,1,3,5)==>(2,3,5)
**矩阵最高维为N，则矩阵其实是由n个N-1维的矩阵构成，如果我们想让m维度之间进行比较，则需要再原索引值-1，如在上述例子中，要比较各个三维之间的值，则需要输入三维高一维的四维的索引值就行，因为4D=3D+3D+3D+...+3D;MAX 4D=MAX[3D 3D 3D ...3D] ==>3D**

# 深拷贝与浅拷贝


a.view() #浅拷贝

```python
#text1.view()相当于一下的效果
text1=np.arange(10)

text2=text1
text2[0]=99

text_view=text1.view()
text_view[0]=88

print(text1)
print(text2)
print(text_view)
```

a.copy() #深拷贝，其效果是上个cell的text2更新值后，不会影响text1，copy之后text1和text2是两个独立互不影响的矩阵，所以修改任何一个对象都不会影响另一个对象

```python
#text1.copy()相当于一下的效果
text1=np.arange(10)

text2=text1
text2[0]=99

text_copy=text1.copy()
text_copy[0]=88

print(text1)
print(text2)
print(text_copy)
```
